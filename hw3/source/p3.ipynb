```python
def memoize(func):

    cache = {}
    
    def memoize_func(*args):
        if args in cache:
            return cache[args]
        
        result = func(*args)
        
        cache[args] = result
        return result 
    
    return memoize_func
```

```python

def CFinite(n, C, S) -> int:
    
    if len(C) != len(S):
        print("ERROR")
        return []
    
    d = len(C)

    if n < d:
        return S[n]
    
    return sum( [C[i-1] * CFinite(n-i, C, S) for i in range(1,d+1)] )


```

```python

fib_seq = [CFinite(n, [1,1], [0,1]) for n in range(0, 21)]
fib_seq_square = [ x**2 for x in  fib_seq ]
```
