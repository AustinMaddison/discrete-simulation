Ajarn prposed, Aproximate square root 2 using probability of x^2 <= 2 between 0 to 2 interval

```python
def square_root_two_approx(n = 1000000, x=2):
    xs = np.random.ranf(n) * x
    ys = xs**2
    count = ys <= x
    return np.mean(count) * x
    
print(square_root_two_approx(x=2))
```

Fib

```python
cache = {}
def fib(n):
    
    if n in cache:
        # print("hit")
        return cache[n]
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    cache[n] = fib(n - 1) + fib(n - 2)
    return cache[n]

leading_digit =  np.vectorize(lambda _: int(str(_)[0]))


fib_vec = np.vectorize(lambda _: fib(_))

print(fib_vec(1))

def benfords(n = 30):
    x = np.arange(1, n+1, 1, dtype=np.uint32)
    print(x)
    
    x = fib_vec(x)
    
    
    x = leading_digit(x)
    # print(x)
    # plt.hist(x, density=True, bins=np.arange(1, 10, 1))
    
    print(x)
    hist, edge = np.histogram(x, density=True, bins=np.arange(1, 11, 1))

    print(hist.size)
    plt.figure(figsize=(6, 4))
    plt.grid()
    plt.plot(np.arange(1, 10, 1), hist, color='k')
    
    print(hist)
    
    print(np.sum(hist))
    
    
benfords()
```
